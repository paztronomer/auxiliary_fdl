```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def quick_plot(x1, tt):
    plt.scatter(np.log10(1/x1["r"]), np.log10(x1["alpha"]), c="orange", s=15, marker="D", alpha=0.5, label=r"$\alpha$")
    plt.scatter(np.log10(1/x1["r"]), np.log10(x1["N"]), c="forestgreen", s=5, marker="*", alpha=0.5, label="D")
    plt.xlabel(r"$log_{10}$(1/r)")
    plt.ylabel(r"$log_{10}${N, $\alpha$}")
    plt.legend(loc="upper center")
    plt.title(tt)
    plt.show()
```

```python
DEPTH = 0
path = "fdl_arithmetic"
for root,dirs,files in os.walk(path):
    if root.count(os.sep) >= DEPTH:
        del dirs[:]
    for index, item in enumerate(files):
        if (("kplr" in item) or ("g_" in item)):
            fnm = os.path.join(root,item)
            x = np.load(fnm)
            quick_plot(x, item)
```
